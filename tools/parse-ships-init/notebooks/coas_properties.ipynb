```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
shipsdDf = pd.read_csv('../ship_init_COAS.csv', index_col=0)
shipsdDf
```

```python
for key in shipsdDf.keys()[5:-1]:
    fig, ax = plt.subplots()
    ax.scatter(shipsdDf['Class'], shipsdDf[key])
    ax.set_xlabel('Class')
    ax.set_ylabel(key)
```
